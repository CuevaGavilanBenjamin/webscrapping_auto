```python
import pandas as pd
```

```python
df=pd.read_csv("../TRANSFORM/reporte_consolidado.csv")
```

```python
# Agrupar por fecha y calcular estadísticas (promedio, min, max, etc.)
reporte = df.groupby('Fecha').agg({
    'compra': ['mean', 'min', 'max'],
    'venta': ['mean', 'min', 'max']
}).reset_index()

print(reporte)
```

```python
import matplotlib.pyplot as plt

# Filtrar el DataFrame por la moneda específica
df_dolar = df[df['moneda'] == 'Dólar de N.A.']

# Crear un gráfico de líneas para la compra y venta del Dólar de N.A.
plt.figure(figsize=(10, 6))
plt.plot(df_dolar['Fecha'], df_dolar['compra'], label='Compra', color='green')
plt.plot(df_dolar['Fecha'], df_dolar['venta'], label='Venta', color='red')
plt.xlabel('Fecha')
plt.ylabel('Valor (S/)')
plt.title('Evolución del Tipo de Cambio - Dólar de N.A.')
plt.xticks(rotation=45)
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Lista de monedas que quieres graficar
monedas = ['Dólar de N.A.', 'Euro', 'Libra Esterlina']

# Crear el gráfico
plt.figure(figsize=(12, 8))

for moneda in monedas:
    df_moneda = df[df['moneda'] == moneda]
    plt.plot(df_moneda['Fecha'], df_moneda['compra'], label=f'Compra - {moneda}')
    plt.plot(df_moneda['Fecha'], df_moneda['venta'], label=f'Venta - {moneda}')

plt.xlabel('Fecha')
plt.ylabel('Valor (S/)')
plt.title('Evolución del Tipo de Cambio por Moneda')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
